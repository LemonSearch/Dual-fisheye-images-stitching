```python
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import scipy
import cv2
```

Create a vignetting mask to remove the vignetting effects

```python
def create_vignette(image, strength=0.4, protect=0.2):
    # create a vignette mask in the same size as image
    mask = np.zeros_like(image)

    # calculate the center
    center_x = image.shape[1] // 2
    center_y = image.shape[0] // 2

    # calculate the radius
    r = np.sqrt((center_x ** 2) + (center_y ** 2))

    # create visible area
    cv2.circle(mask, (center_x, center_y), image.shape[0]//2, (255,255,255), -1)  # Draw the circle on original image

    # apply vignette
    for y in range(image.shape[0]):
        for x in range(image.shape[1]):
            distance = np.sqrt(((x - center_x) ** 2) + ((y - center_y) ** 2))

            vignette_strength = min(1 - (distance / r) * strength + protect, 1)
            mask[y, x] = mask[y, x] * (1 - vignette_strength)

    return mask
```

To show images in Notebook

```python
def show_img(img):
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img_rgb)
    plt.axis('off')  # 关闭坐标轴
    plt.show()
```

### show fisheye images

```python
im = cv2.imread("F2_3.jpg")
im = im[::4, ::4, :]
ima = im[:,:im.shape[1]//2]
imb = im[:,im.shape[1]//2:]


mask = create_vignette(ima)


ima = np.clip(ima/2+mask/2, 0, 127.5) * 2
ima = ima.astype(np.uint8)
imb = np.clip(imb/2+mask/2, 0, 127.5) * 2
imb = imb.astype(np.uint8)

# cv2.imwrite("fisheye_image.jpg", imb)
show_img(ima)
show_img(imb)

```

### Unwarp fisheye images

```python
def get_grid(im):
    h, w = im.shape[:2]
    x = np.arange(w)
    y = np.arange(h)
    x_prime = (2 * x / w - 1)
    y_prime = (2 * y / h - 1)
    grid1, grid2 = np.meshgrid(y_prime, x_prime)
    return grid1.ravel(), grid2.ravel()
    

def fisheye2d_to_vector3d(aperture, k, x, y):
    r = np.linalg.norm(np.stack([x, y]), axis=0)
    aperture = np.deg2rad(aperture)

    try:
      if -1 <= k < 0:
          f = k / np.sin(k*aperture/2)
          phi = np.arcsin(np.clip(r * k / f, -1, 1)) / k
      elif k == 0:
          f = 2 / aperture
          phi = r / f
      elif 0 < k <= 1:
          f = k / np.tan(k*aperture/2)
          phi = np.arctan(r * k / f) / k
      else:
          raise ValueError("k must be in the range [-1, 1]!")
    except ValueError as error:
      print(error)

    theta = np.arctan2(y, x)
    return r, phi, theta


def vector3d_to_latlon(r, phi, theta):
    P_y = r * np.cos(phi)
    P_z = r * np.sin(phi) * np.sin(theta)
    P_x = r * np.sin(phi) * np.cos(theta)
    longitude = np.arctan2(P_y, P_x)
    latitude = np.arctan2(P_z, np.sqrt(P_x**2 + P_y**2))
    return longitude, latitude


def latlon_to_equirectangular(longitude, latitude):
    x = longitude / np.pi
    y =  2 * latitude / np.pi
    x = ((x + 1) % 2) -1
    y = ((y + 1) % 2) -1
    return x, y


```

```python
def fisheye_unwarp(fisheye_img, output_height=500, lambda_0=-np.pi/2, aperture=190, fisheye_factor=0):
  y_grid, x_grid = get_grid(fisheye_img)

  r, phi, theta = fisheye2d_to_vector3d(aperture, fisheye_factor, y_grid, x_grid)

  ind = r<1
  longitude, latitude = vector3d_to_latlon(r[ind], phi[ind], theta[ind])
  longitude = longitude + lambda_0

  equirectangular_x, equirectangular_y = latlon_to_equirectangular(longitude, latitude)

  output_width = int(output_height/aperture*360)
    
  render_grid = tuple(np.mgrid[-1:1:complex(0, output_height), -1:1:complex(0, output_width)])

  output = scipy.interpolate.griddata(
    np.stack([equirectangular_y, equirectangular_x]).T, fisheye_img.reshape(-1, 3)[ind], render_grid, method='nearest')

  return output.astype(np.uint8)
```

```python
aperture = 199
fisheye_factor=-0.5

im_out_a = fisheye_unwarp(ima, output_height=2000, aperture=aperture, fisheye_factor=fisheye_factor)
im_out_b = fisheye_unwarp(imb, output_height=2000, aperture=aperture, fisheye_factor=fisheye_factor)

# Add lines to indicate the overlap area
x_size = im_out_a.shape[1]
line_color = np.array([255, 255, 255])

deg90 = int(x_size/4)
deg270 = int(3* x_size/4)
left_area_start = int(deg90-(aperture - 180)*x_size/360/2)
left_area_end = int(deg90+(aperture - 180)*x_size/360/2)
right_area_start = int(deg270-(aperture - 180)*x_size/360/2)
right_area_end = int(deg270+(aperture - 180)*x_size/360/2)

im_out_a[:,left_area_start] = line_color
im_out_a[:,left_area_end] = line_color
im_out_a[:,right_area_start] = line_color
im_out_a[:,right_area_end] = line_color
im_out_b[:,left_area_start] = line_color
im_out_b[:,left_area_end] = line_color
im_out_b[:,right_area_start] = line_color
im_out_b[:,right_area_end] = line_color

show_img(im_out_a)
show_img(im_out_b)
```

### Show overlap area

```python
def overlap_area(unwarp_img, aperture=190):
    x_size = unwarp_img.shape[1]
    
    deg90 = int(x_size/4)
    deg270 = int(3* x_size/4)
    
    left_area_start = int(deg90-(aperture - 180)*x_size/360/2)
    left_area_end = int(deg90+(aperture - 180)*x_size/360/2)
    right_area_start = int(deg270-(aperture - 180)*x_size/360/2)
    right_area_end = int(deg270+(aperture - 180)*x_size/360/2)
    
    overlap_l = unwarp_img[:, left_area_start:left_area_end]
    overlap_r = unwarp_img[:, right_area_start:right_area_end]

    return overlap_l, overlap_r
```

```python
ima_overlap_l, ima_overlap_r = overlap_area(im_out_a, aperture=aperture)
imb_overlap_l, imb_overlap_r = overlap_area(im_out_b, aperture=aperture)

show_img(ima_overlap_r)
show_img(imb_overlap_l)
```

### Keypoints matching

```python
import cv2

# detect ORB keypoints
orb = cv2.ORB_create(nfeatures=2000, scaleFactor=1.2, nlevels=15)
keypoints_ima_r, descriptors_ima_r = orb.detectAndCompute(ima_overlap_r, None)
keypoints_imb_l, descriptors_imb_l = orb.detectAndCompute(imb_overlap_l, None)

print("There are {} keypoints in the first image.".format(len(keypoints_ima_r)))
print("There are {} keypoints in the second image.".format(len(keypoints_imb_l)))

# Match the key points
bf = cv2.BFMatcher_create(cv2.NORM_HAMMING, crossCheck=True)
matches = bf.match(descriptors_ima_r, descriptors_imb_l)

# select good match (set maximum distance to reduce "X" matching)
max_dis = 0.2 * ima_overlap_r.shape[1]
good_matches = []
for match in matches:
    img1_idx = match.queryIdx
    img2_idx = match.trainIdx
    (x1, y1) = keypoints_ima_r[img1_idx].pt
    (x2, y2) = keypoints_imb_l[img2_idx].pt
    
    if np.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2) < max_dis:
        good_matches.append(match)

print("After select, there are {} good matches.".format(len(good_matches)))

# draw matches
matchImg = cv2.drawMatches(ima_overlap_r, keypoints_ima_r, imb_overlap_l, keypoints_imb_l, good_matches, None)

# draw keypoints
ima_overlap_r_kp = cv2.drawKeypoints(ima_overlap_r, keypoints_ima_r, None)
imb_overlap_l_kp = cv2.drawKeypoints(imb_overlap_l, keypoints_imb_l, None)

# display images
show_img(ima_overlap_r_kp)
show_img(imb_overlap_l_kp)
show_img(matchImg)
cv2.imwrite("match_img.jpg", matchImg)

```
